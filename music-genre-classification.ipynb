```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization, Dropout, GlobalAveragePooling2D, Dense
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint
```

```python
def normalize(x):
    m = np.mean(x)
    s = np.std(x)
    
    return (x - m) / s
```

```python
X = np.load("./data/X.npy")
Y = np.load("./data/Y.npy")

X = normalize(X)
```

```python
NAME = "mgc-cnn"

tensorboard = TensorBoard(log_dir=f"./logs/{NAME}")
```

```python
model = Sequential()

model.add(Conv2D(16, (5,5), padding="same", activation="elu", input_shape=X.shape[1:]))
model.add(MaxPooling2D())

model.add(Conv2D(32, (5,5), padding="same", activation="elu"))
model.add(MaxPooling2D())
model.add(BatchNormalization())
model.add(Dropout(0.60))
          
model.add(Conv2D(64, (5,5), padding="same", activation="elu"))
model.add(MaxPooling2D())

model.add(Conv2D(64, (5,5), padding="same", activation="elu"))
model.add(MaxPooling2D())
model.add(BatchNormalization())
model.add(Dropout(0.60))

model.add(Conv2D(128, (5,5), padding="same", activation="elu"))
model.add(MaxPooling2D())
model.add(Dropout(0.60))

model.add(Conv2D(256, (5,5), padding="same", activation="elu"))

model.add(Conv2D(256, (1,1), padding="same", activation="elu"))
model.add(BatchNormalization())
          
model.add(GlobalAveragePooling2D())
model.add(BatchNormalization())

model.add(Dense(256, activation="elu"))
model.add(BatchNormalization())
model.add(Dropout(0.60))

model.add(Dense(8, activation="softmax"))
          
model.compile(optimizer="adam", 
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

model.summary()
```

```python
filepath = "./models/weights.{epoch:03d}-{val_acc:.2f}.model"
checkpoint = ModelCheckpoint(filepath, "val_acc", 0, True)

model.fit(X, Y, batch_size=32, epochs=500, validation_split=0.05, callbacks=[tensorboard, checkpoint])
```
